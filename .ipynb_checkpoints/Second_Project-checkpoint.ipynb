```python
#Import needed libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read dataset
df = pd.read_csv('Crime_Data.csv')
```

```python
# to check if the dataset is loaded successfully
df.head()
```

```python
# print the shape of dataset
df.shape
```

```python
#print info of dataset
df.info()
```

```python
#print describe of dataset
df.describe()
```

```python
# check missing values in each column
df.isna().sum() 
```

```python
#print rows contain messing values
df[df.isna()]
```

```python
#for the missing values in date of crimes we just drop them
df.dropna(subset=['Occurred Date'], inplace=True)
```

```python
#print the cumber of missing values in each column
df.isna().sum() 
```

```python
# fill the missing values in other columns by using mode() function
x = df['Crime Subcategory'].mode()[0]
y = df['Sector'].mode()[0]
z = df['Precinct'].mode()[0]
w = df['Beat'].mode()[0]
df['Crime Subcategory'].fillna(x, inplace = True)
df['Sector'].fillna(y, inplace = True)
df['Precinct'].fillna(z, inplace = True)
df['Beat'].fillna(w, inplace = True)
```

```python
#final check of missing values
df.isna().sum() 
```

```python
# print the number of duplicated rows and duplicated rows
num_duplicates = df.duplicated().sum()
print(f'The dataset contains {num_duplicates} duplicated rows.')
df[df.duplicated()]
```

```python
# drop duplicated rows
df.drop_duplicates(inplace=True, keep='first')
```

```python
# final check of duplicated values
num_duplicates = df.duplicated().sum()
print(f'The dataset contains {num_duplicates} duplicated rows.')
```

```python
# drop irrelevant columns
irrelevant_columns = ['Report Number','Precinct', 'Sector', 'Beat', 'Neighborhood']
df = df.drop(columns=irrelevant_columns)
```

```python
# number of rows and columns after data cleaning
print('Row count is:', len(df.index))
print('Columns count is:', len(df.columns))
```

```python
df.head()
```

```python
# reformat of reported time of crime to %H%M, then we saved the result to new CSV file
# Preprocess the "Time" column

df['Reported Time'] = df['Reported Time'].astype(str).str.split('.').str[0]  #preprocess the "Reported Time" column, remove decimal points
df['Reported Time'] = df['Reported Time'].astype(int) # convert the "Time" column to integer format
df['Reported Time'] = pd.to_datetime(df['Reported Time'], format='%H%M', errors='coerce') # convert the "Reported Time" column to datetime format
df['Reported Time'] = df['Reported Time'].dt.strftime('%H:%M') # convert the "Reported Time" column to the desired format "HH:MM"

df['Occurred Time'] = df['Occurred Time'].astype(str).str.split('.').str[0]  #preprocess the "Reported Time" column, remove decimal points
df['Occurred Time'] = df['Occurred Time'].astype(int) # convert the "Time" column to integer format
df['Occurred Time'] = pd.to_datetime(df['Occurred Time'], format='%H%M', errors='coerce') # convert the "Reported Time" column to datetime format
df['Occurred Time'] = df['Occurred Time'].dt.strftime('%H:%M') # convert the "Reported Time" column to the desired format "HH:MM"
df
```

```python
df.dropna(subset=['Occurred Time'], inplace=True)
df
```

```python
#What is the distribution of crime occurrences throughout the day?
```

```python
# Convert 'Occurred Time' column to string format
df['Reported Time'] = df['Reported Time'].astype(str)

# Pad single-digit time values with leading zeros
df['Reported Time'] = df['Reported Time'].str.zfill(4)

# Create a new column for the time period
df['Time Period'] = df['Reported Time'].str[:2] + ':00-' + df['Reported Time'].str[:2] + ':59'

# Calculate the count of crimes per time period
period_counts = df['Time Period'].value_counts().sort_index()

# Add missing time periods if midnight is missing
if not period_counts.index.str.contains('00:00').any():
    period_counts['00:00-00:59'] = 0

# Sort the time periods
period_counts = period_counts.sort_index()

# Define the interval for displaying x-axis labels
label_interval = 1

# Plot the line chart
plt.figure(figsize=(8,8))
plt.style.use('ggplot')
plt.step(period_counts.index, period_counts.values)
plt.xlabel('Time Period',fontsize = 10)
plt.ylabel('Number of Crimes',fontsize = 10)
plt.title('Distribution of Crime Reported by Time Period',fontsize = 20)
plt.legend(['Number of Crimes'], shadow = True, loc = 0) # location (0 = best)

# Set the x-axis label display interval
plt.xticks(range(0, len(period_counts.index), label_interval), rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
NARCOTIC_data = df[df['Crime Subcategory'] == 'NARCOTIC']
NARCOTIC_data2 = NARCOTIC_data['Primary Offense Description'].value_counts().head(10)
NARCOTIC_data2_df = pd.DataFrame({'Primary Offense Description': NARCOTIC_data2.index, 'Number of Reported Crimes': NARCOTIC_data2.values})
plt.figure(figsize=(8, 8))
sns.catplot(data=NARCOTIC_data2_df, x='Primary Offense Description', y='Number of Reported Crimes', kind='bar')
plt.xlabel('Primary Offense Description', fontsize=10)
plt.ylabel('Number of Reported Crimes', fontsize=10)
plt.title('Top 10 Reported Crimes for NARCOTIC Category', fontsize=20)
plt.legend(['Number of Crimes'], shadow = True, loc = 0) # location (0 = best)
plt.xticks(rotation=90)
plt.show()
```

```python
# Filter the DataFrame to extract data related to the 'WEAPON' crime subcategory
Weapon_data = df[df['Crime Subcategory'] == 'WEAPON']
# Create a new DataFrame to store the counts of each unique 'Primary Offense Description' in the weapon data
Weapon_data2 = pd.DataFrame(Weapon_data['Primary Offense Description'].value_counts().reset_index())
# Set up the figure size for the pie chart
plt.figure(figsize=(6, 6))
# Create a pie chart to represent the distribution of weapon crime types
plt.pie(Weapon_data2['Primary Offense Description'], labels=Weapon_data2['index'], autopct='%1.2f%%', startangle=90, shadow=False, radius=1, explode=[0.2, 0, 0, 0])
# Set the title for the pie chart
plt.title('Weapon Crime Types', loc='center', fontsize=16)
plt.show()
```

```python
#Monthly Trends of Theft Shoplifting Incidents
```

```python
THEFT_SHOPLIFT_data = df[df['Crime Subcategory'] == 'THEFT-SHOPLIFT']
# Convert 'Occurred Date' column to datetime
THEFT_SHOPLIFT_data['Occurred Date'] = pd.to_datetime(THEFT_SHOPLIFT_data['Occurred Date'])

# Group the data by month and calculate the count of shoplift crimes
THEFT_SHOPLIFT_by_month = THEFT_SHOPLIFT_data['Occurred Date'].dt.month_name().value_counts().sort_index()
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Set the figure size for the plot
plt.figure(figsize=(9, 7))

# Plot the data using markers for each month
plt.plot(months, THEFT_SHOPLIFT_by_month.values, marker='o')

# Add labels to the x-axis and y-axis
plt.xlabel('Month', fontsize=10)
plt.ylabel('Number of Theft Shoplift', fontsize=10)

# Set the title for the plot
plt.title('Number of Theft Shoplift by Month', fontsize=15)

# Add a legend to the plot to indicate the meaning of the data
plt.legend(['Number of Theft Shoplift'], shadow=True, loc=0)  # location (0 = best)

# Rotate the x-axis labels vertically for better readability
plt.xticks(rotation='vertical')

plt.show()
```
